```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Generate synthetic genomic features (100 samples, 10 features)
np.random.seed(42)
X = np.random.rand(100, 10)

# Generate synthetic gene expression levels
# Assume a linear relationship with noise
true_coefficients = np.array([2, -1.5, 3, 0.5, -2, 1.8, -0.7, 1, -1.2, 0.9])
noise = 0.1 * np.random.randn(100)
y = np.dot(X, true_coefficients) + noise

# Create a DataFrame for the dataset
columns = [f"Feature_{i}" for i in range(1, 11)]
df = pd.DataFrame(X, columns=columns)
df['Gene_Expression'] = y

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[columns], df['Gene_Expression'], test_size=0.2, random_state=42)

# Train a simple linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
import matplotlib.pyplot as plt

# Plot predicted vs. actual gene expression levels
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
plt.xlabel('Actual Gene Expression')
plt.ylabel('Predicted Gene Expression')
plt.title('Predicted vs. Actual Gene Expression Levels')
plt.show()
```

```python
# Get input from the user for genomic features
user_input = []
for i in range(1, 11):
    feature_value = float(input(f"Enter value for Feature_{i}: "))
    user_input.append(feature_value)

# Convert user input to a DataFrame
user_df = pd.DataFrame([user_input], columns=columns)

# Make prediction using the trained model
user_pred = model.predict(user_df)

# Print the predicted gene expression level
print(f"Predicted Gene Expression Level: {user_pred[0]}")

# Explain the result to the user
print("Explanation:")
print("- A linear regression model was trained to predict gene expression levels based on genomic features.")
print("- The model used coefficients learned from the training data to make predictions.")
print("- Based on the genomic features you provided, the model predicts a gene expression level of approximately", user_pred[0])

```
